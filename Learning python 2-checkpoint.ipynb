```python
# print 5

Number = 5
while Number < 6:
    print(Number)
    Number+=1

# This means, print 5 in as much as 5 is less than 6. So since 5 is less 6 the condition evaluates to True
```

```python
# break: If you wants to break the loop halfway when certain conditions are met, use break keyword 
# break the loop when it gets to the 12th number 
Number = 7
while Number < 19:
    print(Number)
    if Number == 12:
        break
    Number += 1
```

```python
# Numpy Differences: Differences means subtracting an element from the successive one. This is done using diff()
# From the array below find the differences

My_List = np.array([7.8,5.8,8.4,3.2,6.1,7.0,5.4,9.3,5.8])
The_Diff = np.diff(My_List)
print(The_Diff)
```

```python
''' Numpy LCM(Lowest Common Multiple): LCM is the least number that when multiplied 2 numbers differently
they will have a common answer to. The number that both has common multiplicate number with''''
# For example 12 is the least number that after multiplying 4 and 6 with diff numbers that both will have same number with

import numpy as np
First_Number = 4
Second_Number = 6
Both_Numbers = np.lcm(First_Number,Second_Number)
print(Both_Numbers)
```

```python
# Finding LCM in Arrays. This could be done using reduce(): 
My_Numbers = np.array([7,5,8,3,6,7,5,9,8])
The_LCM = np.lcm.reduce(My_Numbers)
print(The_LCM)
```

```python
'''Numpy Greatest Common Denominator: This is also known as Highest Common Factor. It is the biggest number that is 
a common factor of both of the numbers or the highest number that all values can be divided with '''
# THis is done using gcd()
import numpy as np
Numba1 = 4
Number2 = 6
Both_Numbas = np.gcd(Numba1, Number2)
print(Both_Numbas)
```

```python
# Finding GCD in Arrays. This could be done using reduce(): 
My_Numbers = np.array([7,5,8,3,6,7,5,9,8])
The_LCM = np.gcd.reduce(My_Numbers)
print(The_LCM)
```

```python
# Set Operations: A set is a collection of unique elements
# unique(): is used in creating a set in numpy
# create a set from the array below 

import numpy as np
Set_Array = np.array([1,1,1,2,3,4,5,5,6,7])
The_Set_Creation = np.unique(Set_Array)
print(The_Set_Creation)
```

```python
# Finding the union of a set. This means finding the unique values in 2 arrays. this is done using union1d()
import numpy as np
Set_1 = [1,5,3,9,4]
Set_2 = [8,4,7,2,5]
The_Union = np.union1d(Set_1,Set_2)
print(The_Union)
```

```python
# finding the intersection of a set: This is to find only the values that are present in both arrays.
# this is done using the intersect1d()
# find the intersection of the following two set arrays

import numpy as np
Intersect_Set_1 = [1,5,3,9,4]
Intersect_Set_2 = [8,4,7,2,5]
The_Intersect = np.intersect1d(Intersect_Set_1, Intersect_Set_2)
print(The_Intersect)
```

```python
# Finding difference: This is to find the values in the first array or set that's not in the second set
# this is done using setdiff1d()
# find the difference of the following two set arrays

import numpy as np
Diff_Set_1 = np.array([1,5,3,9,4])
Diff_Set_2 = np.array([8,4,7,2,5])
The_Difference = np.setdiff1d(Diff_Set_1, Diff_Set_2)
print(The_Difference)
```

```python
# Finding symmetric difference: This is to find the values that are not present in both set or array
# this is done using setxor1d()
# find the symetric difference of the following two set arrays

import numpy as np
Diff_Set_1 = np.array([1,5,3,9,4])
Diff_Set_2 = np.array([8,4,7,2,5])
The_Difference = np.setxor1d(Diff_Set_1, Diff_Set_2)
print(The_Difference)
```
